```python
import pandas as pd
```

```python
df = pd.read_csv("Diseases_data.csv", error_bad_lines=False)
```

```python
df = df.fillna(df.mean())
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
fig= plt.subplots(nrows=1, ncols=1, figsize=(15, 9))
plt.scatter(df["Year"],df["Acute Diarrhoeal Diseases - Cases"],label="Acute Diarrhoeal", marker="o")
plt.scatter(df["Year"],df["Malaria - Cases"],label="Malaria", marker="^")
plt.scatter(df["Year"],df["Japanese Encephalitis - Cases"],label="Japanese Encephalitis", marker="s")
plt.scatter(df["Year"],df["Acute Respiaratory Infection - Cases"],label="Acute Respiaratory Infection", marker="x")
plt.scatter(df["Year"],df["Viral Hepatitis - Cases"],label="Viral Hepatitis", marker="*")
plt.ylabel("No. of Cases", fontsize= "12")
plt.xlabel("Disease", fontsize= "12")
plt.title("Year vs Number of Cases due to Diseases", fontsize= "15")
plt.legend()
```

```python
data = [df["Acute Diarrhoeal Diseases - Deaths"],df["Malaria - Deaths"], df["Japanese Encephalitis - Deaths"],df["Acute Respiaratory Infection - Deaths"],df["Viral Hepatitis - Deaths"]]
      
fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(15, 9))
bplot1 = ax1.boxplot(data,vert=True,patch_artist=True, labels = ['Acute Diarrhoeal', 'Malaria', 'Japanese Encephalitis', 'Acute Respiaratory Infection', 'Viral Hepatitis']) 
                     
                   
# Creating axes instance
colors = ['pink', 'lightblue', 'lightgreen', 'lightblue', 'lightgreen']

for patch,color in zip(bplot1['boxes'], colors):
    patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Disease')
    ax.set_ylabel('No. of deaths')
    ax.set_title("Year vs Number of deaths due to Diseases", fontsize= "15")

plt.show()
# Creating plot
bp = ax.boxplot(data)

# show plot
plt.show()
```
